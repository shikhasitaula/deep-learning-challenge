## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
# import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
cleaned_application_df = application_df.drop(columns=["EIN", "NAME"])
```

```python
# Determine the number of unique values in each column.
unique_value_counts = cleaned_application_df.nunique()

# Print the result
print(unique_value_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = cleaned_application_df['APPLICATION_TYPE'].value_counts()

# Print the value counts
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_type_counts = cleaned_application_df['APPLICATION_TYPE'].value_counts()

# Choose a cutoff value 
cutoff_value = 3


# Replace in dataframe
for app in application_type_counts:
    cleaned_application_df['APPLICATION_TYPE'] = cleaned_application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
cleaned_application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = cleaned_application_df['CLASSIFICATION'].value_counts()

# Print the value counts
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_greater_than_1 = classification_counts[classification_counts > 1]
print(classification_value_greater_than_1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = cleaned_application_df['CLASSIFICATION'].value_counts()

# Replace in dataframe
for cls in classifications_to_replace:
    cleaned_application_df['CLASSIFICATION'] = cleaned_application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
cleaned_application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
cleaned_application_df = pd.get_dummies(cleaned_application_df,dtype=float)
cleaned_application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = cleaned_application_df.drop('IS_SUCCESSFUL', axis=1).values

# Target (y) is the target column
y = cleaned_application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
# Split the data into a training set (70%) and a testing set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size=0.25,, random_state=42)

# Print the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
